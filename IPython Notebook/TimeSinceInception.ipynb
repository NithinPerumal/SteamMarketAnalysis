```python
import json;
import pandas as pd;
import matplotlib.pyplot as plt;
import matplotlib;
import datetime as dt
import time
import numpy as np
from sklearn import tree

from pandas.io.json import json_normalize

matplotlib.style.use('ggplot');
%matplotlib inline
```

```python
dfIn = pd.read_csv('itemhistory.csv', dtype={'Craft':str, 'DateTime':dt.datetime, 'DefIndex':np.float64, 'Name':str,
                                            'PriceIndex':str, 'QualInt':np.int64, 'Trade':str, 'currency':str,
                                            'timestamp':np.float64, 'value':np.float64, 'value_high':np.float64})
dfIn['DateTime'] = pd.to_datetime(dfIn['DateTime'])
dfIn = dfIn.set_index(['Name', 'DefIndex', 'QualInt', 'Trade', 'Craft', 'PriceIndex', 'timestamp'])
dfIn.sort_index(inplace = True)
```

```python
dfIn.reset_index(inplace = True)
```

```python
# Initialize the inception column
dfIn['Inception'] = dfIn['DateTime'].min()

# Set inception times for each type of item
names = dfIn['Name'].unique()

for name in names:
    qualities = dfIn[dfIn['Name'] == name]['QualInt'].unique()
    for quality in qualities:
        prices = dfIn[(dfIn['Name'] == name) & (dfIn['QualInt'] == quality)]['PriceIndex'].unique()
        for price in prices:
            inception = dfIn[(dfIn['Name'] == name) & (dfIn['QualInt'] == quality) & (dfIn['PriceIndex'] == price)]['DateTime'].min()
            dfIn['Inception'][(dfIn['Name'] == name) & (dfIn['QualInt'] == quality) & (dfIn['PriceIndex'] == price)] = inception
```

```python
# Set time since inception for each row
dfIn['time_since_inception'] = dfIn['DateTime'] - dfIn['Inception']
dfIn.head(1000)
```
